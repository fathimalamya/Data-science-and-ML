```python
import pandas as pd
df=pd.read_csv(r'C:\Users\97455\Downloads\Wholesale customers data.csv')
df
```

```python
from sklearn.preprocessing import normalize
scaled=normalize(df)
data=pd.DataFrame(scaled,columns=df.columns)
data
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.title('dendrogram')
dend=shc.dendrogram(shc.linkage(scaled,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='complete')
pred=cluster.fit_predict(data)
pred
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(data['Milk'],data['Grocery'],c=cluster.labels_)
```

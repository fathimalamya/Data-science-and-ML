```python

import pandas as  pd
```

```python
df=pd.read_csv(r'C:\Users\97455\Downloads\Iris.csv')
```

```python
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier()
dec.fit(x_train,y_train)
pred=dec.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,pred))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,pred))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(dec,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica' ])
```

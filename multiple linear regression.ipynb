```python
x=df.iloc[:, :-1]
y=df.iloc[:,-1]
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
pred=reg.predict(x_test)
pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,pred)*100
print(score)
```

```python
diff=pd.DataFrame({'actual value':y_test,'predicted value':pred,'diffrence':y_test-pred})
diff
```
